```python
import pandas as pd
import numpy as np
import random

# Load airline and airport codes
airline_codes = ['6E', 'AI', 'SG', 'UK', 'G8', 'I5', 'QP']  # Indian carriers
airport_codes = ['DEL', 'BOM', 'BLR', 'HYD', 'MAA', 'CCU', 'PNQ', 'AMD', 'GOI', 'LKO', 'JAI', 'SXR', 'IXC', 'BHU', 'TRV']

# Probabilities
airline_probs = [0.58, 0.12, 0.15, 0.07, 0.05, 0.02, 0.01]
airport_probs = [0.18, 0.16, 0.14, 0.08, 0.07, 0.08, 0.05, 0.04, 0.04, 0.04, 0.03, 0.03, 0.02, 0.02, 0.02]

# Modifiers
airline_delay_mod = {'6E': -2, 'UK': -5, 'AI': 5, 'SG': 8, 'I5': 4, 'QP': -3, 'G8': 10}
airport_delay_mod = {'DEL': 8, 'BOM': 10, 'BLR': 5}
seasonal_delay_mod = {6: 10, 7: 12, 8: 10, 12: 15, 1: 15}

# Total records
num_flights = 100_000
print(f"Generating {num_flights:,} flight records...")

flights = []
for _ in range(num_flights):
    airline = np.random.choice(airline_codes, p=airline_probs)
    origin, dest = np.random.choice(airport_codes, size=2, replace=False, p=airport_probs)
    
    month = random.randint(1, 12)
    day = random.randint(1, 28)
    hour = random.randint(5, 23)
    sched_departure = hour * 100 + random.randint(0, 59)

    base_delay = np.random.lognormal(2.5, 1) - 5
    delay = (
        base_delay +
        airline_delay_mod.get(airline, 0) +
        airport_delay_mod.get(origin, 0) +
        seasonal_delay_mod.get(month, 0) +
        (hour / 3)
    )

    dep_delay = round(delay * random.uniform(0.4, 0.6), 2)
    arr_delay = round(delay - dep_delay, 2)

    # Cancellation logic
    is_cancelled = 0
    cancel_reason = None
    cancel_prob = 0.01 + (0.05 if month in [7, 1, 12] else 0) + (0.01 if origin in ['DEL', 'BOM'] else 0)

    if random.random() < cancel_prob:
        is_cancelled = 1
        if month in [6, 7, 8, 1, 12]:
            cancel_reason = 'B'  # Weather
        elif origin in ['DEL', 'BOM']:
            cancel_reason = 'C'  # Congestion
        else:
            cancel_reason = 'A'  # Airline Issue
        dep_delay = arr_delay = None

    flights.append({
        'YEAR': 2023,
        'MONTH': month,
        'DAY': day,
        'AIRLINE': airline,
        'ORIGIN_AIRPORT': origin,
        'DESTINATION_AIRPORT': dest,
        'SCHEDULED_DEPARTURE': sched_departure,
        'DEPARTURE_DELAY': dep_delay,
        'ARRIVAL_DELAY': arr_delay,
        'CANCELLED': is_cancelled,
        'CANCELLATION_REASON': cancel_reason
    })

# Create DataFrame & save
df_flights = pd.DataFrame(flights)
df_flights.to_csv('flights.csv', index=False)
print(f"✅ 'flights.csv' saved with {len(df_flights):,} records.")
```
